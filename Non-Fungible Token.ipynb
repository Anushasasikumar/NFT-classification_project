```python
#Converting Categorical to Numerical using label encoders
import pandas as pd
from sklearn.preprocessing import LabelEncoder


# Create a copy for encoding
df_encoded = df.copy()

# Show object (categorical) columns before encoding
print("Before encoding:")
print(df_encoded.select_dtypes(include='object').head())

# Encode categorical columns
label_encoders = {}
for col in df_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = le  # Save encoder

# Show the same columns after encoding
print("After encoding:")
print(df_encoded[list(label_encoders.keys())].head())

# Optional: Confirm data types
print("Data types after encoding:")
print(df_encoded.dtypes)
```

```python
def replace_outliers_iqr(df):
    df_cleaned = df.copy()
    for column in df_cleaned.columns:
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        median = df_cleaned[column].median()

        # Replace outliers with the median
        df_cleaned.loc[(df_cleaned[column] < lower_bound) | (df_cleaned[column] > upper_bound), column] = median
    return df_cleaned

# Apply to top features
X_top = X[selected_features].copy()
X_top_cleaned = replace_outliers_iqr(X_top)# Step 1: After feature selection


# Step 2: Replace outliers using IQR
def replace_outliers_iqr(df):
    df_cleaned = df.copy()
    for column in df_cleaned.columns:
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        median = df_cleaned[column].median()
        df_cleaned.loc[(df_cleaned[column] < lower_bound) | (df_cleaned[column] > upper_bound), column] = median
    return df_cleaned

X_top_cleaned = replace_outliers_iqr(X_top)

# Step 3: Apply log1p transformation
import numpy as np
X_log_transformed = X_top_cleaned.apply(lambda x: np.log1p(x) if np.issubdtype(x.dtype, np.number) and (x >= 0).all() else x)

# Step 4: Visualize
import seaborn as sns
import matplotlib.pyplot as plt

X_melted_log = pd.melt(X_log_transformed, var_name='Feature', value_name='Value')
plt.figure(figsize=(12, 6))
sns.boxplot(x='Feature', y='Value', data=X_melted_log)
plt.title("📦 Boxplot After IQR + Log Transformation")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Redefine X and y to match the SelectKBest context
df = pd.read_csv("nft_return_type.csv")
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['return_type_encoded'] = label_encoder.fit_transform(df['return_type'])
df = df.drop(columns=['return_type', 'Unnamed: 0'], errors='ignore')
X = df.drop(columns=['return_type_encoded'])
X = X.select_dtypes(include=['number']).fillna(0)
y = df['return_type_encoded']

# Use the selected 8 features
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(score_func=f_classif, k=8)
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
X_selected_df = X[selected_features]

# Split the selected features and target
X_train, X_test, y_train, y_test = train_test_split(X_selected_df, y, test_size=0.3, stratify=y, random_state=42)

# Apply LDA
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)

# Evaluation
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred))
print("\n🔍 Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Optional: Confusion matrix heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title("LDA Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# Redefine X and y from your cleaned dataframe
df = pd.read_csv("nft_return_type.csv")
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['return_type_encoded'] = label_encoder.fit_transform(df['return_type'])
df = df.drop(columns=['return_type', 'Unnamed: 0'], errors='ignore')
X = df.drop(columns=['return_type_encoded'])
X = X.select_dtypes(include=['number']).fillna(0)
y = df['return_type_encoded']

# Select the same top 8 features
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(score_func=f_classif, k=8)
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
X_selected_df = X[selected_features]

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected_df, y, test_size=0.3, stratify=y, random_state=42)

# Apply KNN (you can tune n_neighbors)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Evaluate performance
print("✅ KNN Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred))
print("\n🔍 Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Optional: Confusion matrix visualization
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Greens')
plt.title("KNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import RocCurveDisplay
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Binarize the output (multi-class)
y_bin = label_binarize(y, classes=np.unique(y))
n_classes = y_bin.shape[1]

# Train-test split again if needed
X_train, X_test, y_train_bin, y_test_bin = train_test_split(X[selected_features], y_bin, test_size=0.3, random_state=42)

# Re-train models with OneVsRest strategy
models = {
    "LDA": make_pipeline(StandardScaler(), OneVsRestClassifier(LinearDiscriminantAnalysis())),
    "KNN": make_pipeline(StandardScaler(), OneVsRestClassifier(KNeighborsClassifier(n_neighbors=5))),
    "Random Forest": OneVsRestClassifier(RandomForestClassifier(n_estimators=100, random_state=42)),
    "Naive Bayes": OneVsRestClassifier(GaussianNB()),
    "ANN": make_pipeline(StandardScaler(), OneVsRestClassifier(MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)))
}

# Plot setup
plt.figure(figsize=(10, 8))

for name, model in models.items():
    model.fit(X_train, y_train_bin)
    y_score = model.predict_proba(X_test)

    # Compute ROC curve and ROC area for each class, then average
    fpr, tpr, _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC-AUC Curve - Multiclass (OvR)', fontsize=14)
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
